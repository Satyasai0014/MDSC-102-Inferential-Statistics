# $$Visualizing Distributions$$

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

## $$Binomial Distribution$$

```python
#Author: A.Satya Sai

def plot_bin(n,p,x):
    return comb(n,x)*p**x*(1-p)**(n-x)

n = int(input("Enter the x limit:"))
p = float(input("Enter Probability :"))

X = [i for i in range(n+1)]
y = []
for x in X:
    y.append(plot_bin(n,p,x))
    
plt.title(f'Plot of Binomial Random Variable from {X[0]} to {X[-1]}')
plt.xlabel('x', fontsize = 15)
plt.ylabel('f(x)', fontsize = 15)   
plt.plot(X,y,'r*')
plt.bar(X,y)
plt.show()
  
 
```

# $$ Normal Distribution $$  

With x(100 points) using linspace(mu-4sigma , mu+4sigma).



```python
"""Author: A.Satya Sai

    Plotting Normal Distribution 
    
    
    ____________
    Parameters :
    ____________
    
    mu  - Mean
    
    sigma  - Variance 

"""

def normal_plot(mu,sigma):
    
    x= np.linspace(mu-4*sigma,mu+4*sigma,100)
    f= lambda x: (1/sigma*sqrt(2*pi))*e**((-1/2*(((x-mu)/sigma)**2)))
    plt.figure(figsize = (16,9))
    plt.scatter(mu,f(mu),color="red",marker= "*")
    plt.bar(x,f(x))
    #plt.scatter(mu,f(mu),color="green",marker= "*")
    plt.xlabel("x ------------->",fontsize = 15)
    plt.ylabel("f(x)------------->",fontsize = 15)
    plt.title(f"Normal Distribution With Mean -{mu} and Variance -{sigma}")

mu = int(input("Enter the Mean Value : "))
sigma = int(input("Enter the Variance Value : "))
normal_plot(mu,sigma)
```

# $$ Exponential Distribution $$


```python
"""Author: A.Satya Sai

    Plotting Exponential Distribution 
    
    ___________
    
    Parameters : 
    
                Lambda - as  lam
    ___________
    
     
"""
def exponential_plot(lam):
    
    x= np.linspace(0,10,100)
    f= lambda x: lam*np.exp(-lam*x)
    y = []
    for i in range (len(x)):
        y.append(f(x[i]))
    plt.figure(figsize = (16,9))
    plt.plot(x,y,color="g",marker= "*")
    plt.xlabel("x ------------->",fontsize = 15)
    plt.ylabel("f(x)------------->",fontsize = 15)
    plt.title(f"Exponentail Distribution With Lambda - {lam}")

lam = float(input("Enter the Lambda Value : "))

exponential_plot(lam)
```

# $$ Poisson Distribution $$

```python
"""Author: A.Satya Sai

    Plotting Poisson Distribution 
    
    ___________
    
    Parameters : 
    
                Lambda - as  lam - float
                n - int
    ___________
    
    
    
"""
import math

def poi_plot(lam,n):
    
    x= np.arange(0,n)
    f= lambda x: np.exp(-lam)*(lam**x / math.factorial(x))
    y = []
    for i in range (len(x)):
        y.append(f(x[i]))
    plt.figure(figsize = (16,9))
    plt.plot(x,y,color="g",marker= "*")
    plt.xlabel("x ------------->",fontsize = 15)
    plt.ylabel("f(x)------------->",fontsize = 15)
    plt.title(f"Poisson Distribution With Lambda = {lam} and n = {n}")

lam = float(input("Enter the Lambda Value : "))
n = int(input("Enter the n Value : "))

poi_plot(lam,n)

```

# $$ Gamma Distribution $$

```python
"""Author: A.Satya Sai

    Plotting Gamma Distribution 
    
    ___________
    
    Parameters : 
    
                Lambda - as  lam - float
                alpha  - as al - float
    ___________
    
    
    
"""
import math
import scipy
from scipy.special import gamma

def gamma_function(al):
    out=scipy.special.gamma(al)  
    return out


def gamma_plot(al,lam):
    
    x= np.linspace(0,10,200)
    gal = gamma_function(al)
    f= lambda x: ((lam**al)/gal)*(x**(al-1))*np.exp(-lam*x)
    y = []
    for i in range (len(x)):
        y.append(f(x[i]))
        
    plt.figure(figsize = (16,9))
    plt.plot(x,y,color="g",marker= "*")
    plt.xlabel("x ------------->",fontsize = 15)
    plt.ylabel("f(x)------------->",fontsize = 15)
    plt.title(f"Gamma Distribution With  Alpha = {al} & Lambda = {lam}")

al = float(input("Enter the alpha Value(al>0) : "))
lam = float(input("Enter the alpha Value(lam>0) : "))

gamma_plot(al,lam)
```
